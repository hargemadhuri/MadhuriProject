# Exploratory Data Analysis

**Gender**

```python
#creating graph to see count of gender using sns library 
ax=sns.countplot(x="Gender", data=df, color="yellow")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
'''The DataFrame has been grouped by "Gender," and the "Amount" column has been summed within each group. 
The result is sorted in descending order by the "Amount" column, so "Male" has the highest total amount, 
followed by "Female," and then "Other."
as_index=False, which means that the "Gender" column will not be set as the index of the resulting DataFrame.'''
Sales_gender= df.groupby(["Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
Sales_gender
```

```python

sns.barplot(x="Gender",y="Amount", data=Sales_gender, color="orange")
```

From above graphs we can see that most of the buyers are females and purchasing power of females are greater than male

**Age**

```python
df.columns
```

```python
ax=sns.countplot(data=df, x="Age Group", hue="Gender")


for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#total amount vs age group
Sales_age=df.groupby(["Age Group"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)

sns.barplot(x="Age Group", y="Amount", data=Sales_age, color="yellow")
```

From above graphs we can see that most of the buyers are of age group between 26-35 years female

**State**

```python
df.columns
```

```python
#total number of orders from top 10 States
Sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)
Sales_state
```

```python
Sales_state=df.groupby(["State"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False).head(10)

sns.set(rc={"figure.figsize":(10,4)})
sns.barplot(data=Sales_state, x="State", y="Orders", saturation=50)
plt.xticks(rotation=45)
plt.show()
```

```python
#Totalamount/sales from top 10 States
sales_state=df.groupby(["State"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)

sns.set(rc={"figure.figsize":(10,4)})
sns.barplot(data=sales_state, x="State", y="Amount", saturation=50)
plt.xticks(rotation=45)
plt.show()
```

from above graphs we can see that most of the orders  are from Uttar Pradesh, Maharashtra and Karnataka respectively

**Marital Status**

```python
ax=sns.countplot(data=df, x="Marital_Status", color="yellow")

sns.set(rc={"figure.figsize":(4,3)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Status=df.groupby(["Marital_Status", "Gender"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={"figure.figsize":(4,3)})
sns.barplot(data= Sales_Status, x="Marital_Status", y="Amount", hue="Gender", saturation=50)
```

From the above graphs we can see that most of the buyers are married(women) and they have high purchasing power

**Occupation**

```python
sns.set(rc={"figure.figsize":(20,5)})
ax=sns.countplot(data=df, x="Occupation")

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Occu=df.groupby(["Occupation"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=Sales_Occu, x="Occupation", y="Amount")
```

From above graphs we can see that most of the buyers are working in IT, Healthcare and Aviation sector

**Product Category**

```python
sns.set(rc={"figure.figsize":(25,8)})
ax=sns.countplot(data=df, x="Product_Category")
# plt.xticks(rotation=45)
# plt.show()

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
Sales_Product=df.groupby(["Product_Category"], as_index=False)["Amount"].sum().sort_values(by="Amount", ascending=False).head(10)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=Sales_Product, x="Product_Category", y="Amount")
plt.xticks(rotation=45)
plt.show()
```

From above graphs we can see that most of the sold products are from Food, Clothing & Apparel,Electronics & Gadgets

```python
Product_order= df.groupby(["Product_ID"], as_index=False)["Orders"].sum().sort_values(by="Orders", ascending=False)
sns.set(rc={"figure.figsize":(20,5)})
sns.barplot(data=Product_order, x="Product_ID", y="Orders")
plt.show()
```

```python
#top 10 most sold products 
fig1, ax1=plt.subplots(figsize=(12,7))

df.groupby("Product_ID")["Orders"].sum().nlargest(10).sort_values(ascending=False).plot(kind="bar")
```

Married women age group 26-35 yrs from UP, Maharashtra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from food, clothing and Electronics category.

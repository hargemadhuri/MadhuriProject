```python
import random

# Set up the game
words = ['python', 'java', 'javascript', 'ruby', 'php', 'csharp', 'swift']
word = random.choice(words)
letters_guessed = []
guesses_left = 6

# Display the game state
def display_game_state():
    print('Word: ', end='')
    for letter in word:
        if letter in letters_guessed:
            print(letter, end=' ')
        else:
            print('_', end=' ')
    print('\nGuesses left:', guesses_left)
    print('Letters guessed:', ', '.join(letters_guessed))

# Play the game
while True:
    display_game_state()
    guess = input('Guess a letter: ').lower()
    if guess in letters_guessed:
        print('You already guessed that letter!')
    elif guess in word:
        letters_guessed.append(guess)
        if set(word) == set(letters_guessed):
            print('Congratulations, you won!')
            break
    else:
        letters_guessed.append(guess)
        guesses_left -= 1
        if guesses_left == 0:
            print('Sorry, you lost. The word was', word)
            break
```

```python
Copy code
class BankAccount:
    def __init__(self, owner, balance=0):
        self.owner = owner
        self.balance = balance

    def deposit(self, amount):
        self.balance += amount
        print(f"Deposit of {amount} has been made. New balance is {self.balance}")

    def withdraw(self, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            print(f"Withdrawal of {amount} has been made. New balance is {self.balance}")

    def get_balance(self):
        print(f"Current balance is {self.balance}")

    def transfer(self, recipient, amount):
        if amount > self.balance:
            print("Insufficient funds!")
        else:
            self.balance -= amount
            recipient.deposit(amount)
            print(f"Transfer of {amount} has been made to {recipient.owner}. New balance is {self.balance}")

# create two accounts
account1 = BankAccount("John", 1000)
account2 = BankAccount("Jane", 500)

# make a deposit and a withdrawal from account 1
account1.deposit(500)
account1.withdraw(200)

# transfer money from account 1 to account 2
account1.transfer(account2, 300)

# check the balances of both accounts
account1.get_balance()
account2.get_balance()
```

```python
class Book:
    def __init__(self, title, author, isbn):
        self.title = title
        self.author = author
        self.isbn = isbn

class Library:
    def __init__(self):
        self.books = []

    def add_book(self, book):
        self.books.append(book)
        print(f"Book '{book.title}' by {book.author} with ISBN {book.isbn} has been added to the library.")

    def remove_book(self, isbn):
        for book in self.books:
            if book.isbn == isbn:
                self.books.remove(book)
                print(f"Book with ISBN {isbn} has been removed from the library.")
                return
        print(f"Book with ISBN {isbn} was not found in the library.")

    def view_books(self):
        if not self.books:
            print("The library is currently empty.")
        else:
            print("The following books are currently in the library:")
            for book in self.books:
                print(f"'{book.title}' by {book.author} with ISBN {book.isbn}")

# create a library
my_library = Library()

# add some books to the library
book1 = Book("The Great Gatsby", "F. Scott Fitzgerald", "978-0743273565")
my_library.add_book(book1)

book2 = Book("To Kill a Mockingbird", "Harper Lee", "978-0446310789")
my_library.add_book(book2)

# view the books in the library
my_library.view_books()

# remove a book from the library
my_library.remove_book("978-0743273565")

# view the books in the library again
my_library.view_books()
```
